7.	Crea una matriz de subgráficas de 2 x 3 con scatter plots del puntaje (score) versus los 6 factores, para determinar qué factor influye más en la evaluación.

```python
scat = hr
scat.drop(columns= scat[['Overall rank','Country or region']],inplace = True)
scat

```

```python

# Create subplots for each column
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Plot scatter plots for each column
for i, col in enumerate(scat.columns[1:]):  # Start from the second column, excluding 'Score'
    sns.scatterplot(x='Score', y=col, data=scat, ax=axes[i],color='mediumseagreen')
    axes[i].set_xlabel('Score')
    axes[i].set_ylabel(col)
    axes[i].set_title(f'Scatter plot of Score vs {col}')

# Adjust layout
plt.tight_layout()
plt.show()

```

Comprueba lo anterior con un heatmap donde incluyas los índices de correlación. 

```python


# Compute the correlation matrix
correlation_matrix = scat.corr()

# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='BrBG', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

8.	Combina con el dataframe metada (Metadata.csv) para graficar la felicidad promedio por región. 

```python
meta = pd.read_csv('Metadata.csv')
meta.head(4)
```

```python

df = hr
df.columns
```

```python
table1 = df[['Country or region','Score',]]
table1
```

```python
nan_count = table1.isna().sum()
nan_count
```

```python
table2 = meta[['TableName','Region']]
table2
```

```python
merged_table = pd.merge(table1, table2, how='left', left_on='Country or region', right_on='TableName')
merged_table = merged_table.rename(columns= {'Country or region':'Country_t1','TableName':'Country_t2'})
merged_table
```

```python
empty_region = merged_table.isna().sum()
empty_region
```

```python
nan_rows = merged_table[merged_table.isna().any(axis=1)]

print("Rows with NaN values:")
nan_rows
```

```python
happyness = merged_table[['Score','Region']]
hgraph = happyness.groupby('Region').mean()
lastgraph = hgraph.T
```

```python


# Plot the DataFrame with bars using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=lastgraph)
plt.title('World Happiness Report')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.xticks(rotation=45)  # Rotate x-labels for better readability
plt.show()
```
